## 1. Load Filtered Data
We load the filtered dataset from the EDA step, which contains 30,936 expressed genes.

```python
# Load filtered dataset from EDA
df_counts_filtered = pd.read_csv('../data/counts_filtered.csv', index_col=0)

print(f"Dataset loaded: {df_counts_filtered.shape[0]} genes x {df_counts_filtered.shape[1]} samples")
df_counts_filtered.head()
```

## 2. CPM Normalization
We normalize raw counts using CPM (Counts Per Million) to make samples 
comparable regardless of sequencing depth.

```python
# CPM Normalization
def normalize_cpm(df):
    return df.div(df.sum(axis=0), axis=1) * 1_000_000

df_cpm = normalize_cpm(df_counts_filtered)

print("CPM Normalization complete ✓")
print(f"\nTotal counts per sample (should be ~1,000,000):")
print(df_cpm.sum(axis=0).round(0))
df_cpm.head()
```

### 3. Log Transformation

To stabilize variance and reduce the impact of extreme values, we apply a log2 transformation to the CPM-normalized data.

A small constant (+1) is added to avoid issues with zero values.

```python
# Log2 transformation
df_log_cpm = np.log2(df_cpm + 1)

print("Log transformation complete ✓")
df_log_cpm.head()
```
